# Data Visualization with Pandas and Matplotlib

This notebook reads a CSV file, performs basic data exploration, and plots a graph using Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/oasis_longitudinal.csv')

df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Age'], df['MMSE'], alpha=0.6)
plt.title('MMSE Scores by Age')
plt.xlabel('Age')
plt.ylabel('MMSE')
plt.grid(True)
plt.tight_layout()
plt.show()
```
